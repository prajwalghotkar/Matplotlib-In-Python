```python
# Pie Chart-autopact with one decimal point.......
hours=[5,8,3,4]
activity=['Sleep','Work','Eat','Play']
col=['g','r','c','y']
plt.pie(hours,labels=activity,colors=col,startangle=90,shadow=True,explode=(0,0,0,0.1),autopct='%1.1f%%')
plt.title("Pie Chart")
plt.axis('equal')
plt.show()
```
